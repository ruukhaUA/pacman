# 15.5.1. Task 1 — Improve the evaluation function with new heuristics

Since heuristics 1 and 2 were already implemented, we jump straight to the definition of the following ones.

## Heuristic 3 - Food per direction

The quantity of food in a direction will affect its potential selection. This is to encourage the agent to move towards areas richer in food, since it would get stuck in empty areas.



## Heuristic 4 - Number of adjacent walls

Sometimes, the agent would get stuck at corners in a back and forth fashion. Setting a penalization per adjacent wall stops the agent from getting stuck in corners. This penalization doesn't need to be too strong, since it will always evaluate to less than its adjacent spots in the same conditions.



## Heuristic 5 - Distance to solitary food

Leaving behind some pieces of food for no apparent reason would make the agent move great lengths unnecessarily, losing points in the process. Setting a penalization depending on the distance to lonely food (pieces or pair of pieces left alone) forces the agent to collect it when possible, allowing it to go around running from ghosts but not leaving it behind.



## Heuristic 6 - Finish remaining food

There were certain scenarios in which the agent would not finish the food when there wasn't much remaining, since some heuristics awarded more score in these scenarios. To force it to finish all food, when less than 10 remains, it strongly rewards eating the remaining pieces of food.
